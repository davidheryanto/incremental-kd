```python
import os, cPickle, operator, time, random
from collections import defaultdict, OrderedDict

import numpy as np

import theano
import theano.tensor as T
from theano.ifelse import ifelse

import copy
from copy import deepcopy

from logistic_sgd import *
from mlp import *
```

```python
def save_params(file_name, model):
    with open(file_name, 'wb') as out_file:
        cPickle.dump(model.params, out_file)

def set_params(file_name, model):
    with open(file_name, 'rb') as in_file:
        new_params = cPickle.load(in_file)
    for current_p, new_p in zip(model.params, new_params):
        current_p.set_value(new_p.eval())

def reset_params(model, file_name=None):
    if file_name == None:
        file_name = 'model_initial_params.pkl'
    
    print('Before reset:')
    for p in model.params[:2]:
        print(p[0].eval()),
        
    with open(file_name, 'rb') as in_file:
        initial_params = cPickle.load(in_file)
        
    for initial, current in zip(initial_params, model.params):
        current.set_value(initial.eval())
        
    print('\nAfter reset')
    for p in model.params[:2]:
        print(p[0].eval()),
    print('\n')

def save_valid_error(file_name, valid_error):
    with open(file_name, 'wb') as out_file:
        cPickle.dump(valid_error, out_file)

def get_confidence_for_minibatch(index):
    start = index * batch_size
    end = (index + 1) * batch_size
    return np.mean(np.max(get_p_y_given_x(X[start:end]), axis=1))

def should_continue_training(min_valid_error_epoch, epoch_num, diff=10, min_epoch=100):
    if epoch_num < min_epoch:
        return True
    else:
        return False
```

```python
def set_avg_params(paths, model, layer_start=None, layer_end=None):
    if layer_start is None:
        layer_start = 0
    if layer_end is None:
        layer_end = len(model.params)
    
    params = []
    
    for path in paths:
        with open(path, 'rb') as in_file:
            params_loaded = cPickle.load(in_file)
            params.append(params_loaded[layer_start:layer_end])
    
    new_params = []
    for params_tuple in zip(*params):
        total = params_tuple[0].get_value()
        for param in params_tuple[1:]:
            total += param.get_value()
        avg = total / float(len(params))
        new_params.append(avg)
    
    for current_param, new_param in zip(model.params[layer_start:layer_end], new_params):
        # print(current_param.shape.eval(), new_param.shape)
        current_param.set_value(new_param)
```

```python
learning_rate = 0.050
L1_reg = 0
L2_reg = 1e-4
batch_size = 100

nkerns = (80, 40)
n_in = 500
temperature = 3
```

```python
# Use cifar10_david / cifar10_davidheryanto
cifar_path = os.path.join(
    os.getenv('PYLEARN2_DATA_PATH'), 'cifar10_david'
)
datasets = load_cifar10(cifar_path)

train_set_x, train_set_y = datasets[0]
valid_set_x, valid_set_y = datasets[1]
test_set_x, test_set_y = datasets[2]

# compute number of minibatches for training, validation and testing
n_train_batches = train_set_x.get_value(borrow=True).shape[0]
n_valid_batches = valid_set_x.get_value(borrow=True).shape[0]
n_test_batches = test_set_x.get_value(borrow=True).shape[0]
n_train_batches /= batch_size
n_valid_batches /= batch_size
n_test_batches /= batch_size

index = T.lscalar()  # index to a [mini]batch

x = T.matrix('x')  
y = T.ivector('y') 
```

**Train model_1**

```python
print '... building model_1'

rng = numpy.random.RandomState(23455)
random.seed(51665)

model_1 = MLPConv(
    rng=rng,
    model_input=x,
    nkerns=nkerns, 
    n_in=n_in,
    temperature=temperature
)

cost_model_1 = (
    model_1.negative_log_likelihood(y) + \
    L1_reg * model_1.L1 + \
    L2_reg * model_1.L2_sqr
)

test_model_1 = theano.function(
    inputs=[index],
    outputs=model_1.errors(y),
    givens={
        x: test_set_x[index * batch_size: (index + 1) * batch_size],
        y: test_set_y[index * batch_size: (index + 1) * batch_size]
    }
)

validate_model_1 = theano.function(
    inputs=[index],
    outputs=model_1.errors(y),
    givens={
        x: valid_set_x[index * batch_size: (index + 1) * batch_size],
        y: valid_set_y[index * batch_size: (index + 1) * batch_size]
    }
)

grads_model_1 = [T.grad(cost_model_1, param) for param in model_1.params]

updates_model_1 = [
    (param, param - learning_rate * grad)
    for param, grad in zip(model_1.params, grads_model_1)
]

train_model_1 = theano.function(
    inputs=[index],
    outputs=cost_model_1,
    updates=updates_model_1,
    givens={
        x: train_set_x[index * batch_size: (index + 1) * batch_size],
        y: train_set_y[index * batch_size: (index + 1) * batch_size]
    }
)
```

```python
valid_error = []
min_valid_error = np.inf
min_valid_error_epoch = 0
minibatch_indexes = range(int(n_train_batches * 0.25))
epoch_num = 0
print('|Epoch|Valid error|Min (Epoch)|Sec/epoch')
print('|-:|-:|-:|-:')

while should_continue_training(min_valid_error_epoch, epoch_num):
    epoch_num += 1
    start_time = time.time()
    random.shuffle(minibatch_indexes)
    
    for minibatch_index in minibatch_indexes:
        train_model_1(minibatch_index)
        
    valid_loss = np.mean([validate_model_1(i) for i in xrange(n_valid_batches)])
    valid_error.append(valid_loss)
    
    if valid_loss < min_valid_error:
        min_valid_error = valid_loss
        min_valid_error_epoch = epoch_num
        save_params('model_1_best.pkl', model_1)
    
    print('|{:>4}|{:.4f}|{:.4f} ({:>4})|{:>4.1f}').format(
        epoch_num, valid_loss, min_valid_error, min_valid_error_epoch,
        time.time() - start_time
    )
    
# Finish training
save_valid_error('model_1_valid_error.pkl', valid_error)

set_params('./model_1_best.pkl', model_1)
test_loss = np.mean([test_model_1(i) for i in xrange(n_test_batches)])
print('Test error: {}'.format(test_loss))

pylab.rcParams['figure.figsize'] = 8, 6
plt.plot(valid_error)
plt.ylabel('valid error')
plt.xlabel('epoch #')
plt.title('model_1')
```

```python
valid_error = []
min_valid_error = np.inf
min_valid_error_epoch = 0

partition_size = int(n_train_batches / 4.0 / 4.0 )
minibatches_1 = random.sample(range(int(n_train_batches * 0.00), 
                                    int(n_train_batches * 0.25)), partition_size)
minibatches_2 = random.sample(range(int(n_train_batches * 0.25), 
                                    int(n_train_batches * 0.50)), partition_size)
minibatches_3 = random.sample(range(int(n_train_batches * 0.50), 
                                    int(n_train_batches * 0.75)), partition_size)
minibatches_4 = random.sample(range(int(n_train_batches * 0.75), 
                                    int(n_train_batches * 1.00)), partition_size)

minibatch_indexes = copy.deepcopy(minibatches_1)
minibatch_indexes.extend(minibatches_2)
minibatch_indexes.extend(minibatches_3)
minibatch_indexes.extend(minibatches_4)

epoch_num = 0
print('|Epoch|Valid error|Min (Epoch)|Sec/epoch')
print('|-:|-:|-:|-:')

while should_continue_training(min_valid_error_epoch, epoch_num):
    epoch_num += 1
    start_time = time.time()
    random.shuffle(minibatch_indexes)
    
    for minibatch_index in minibatch_indexes:
        train_model_merged_final(minibatch_index)
        
    valid_loss = np.mean([validate_model_merged_final(i) for i in xrange(n_valid_batches)])
    valid_error.append(valid_loss)
    
    if valid_loss < min_valid_error:
        min_valid_error = valid_loss
        min_valid_error_epoch = epoch_num
        save_params('model_merged_final_best.pkl', model_merged_final)
    
    print('|{:>4}|{:.4f}|{:.4f} ({:>4})|{:>4.1f}').format(
        epoch_num, valid_loss, min_valid_error, min_valid_error_epoch,
        time.time() - start_time
    )
    
# Finish training
save_valid_error('model_merged_final_valid_error.pkl', valid_error)

pylab.rcParams['figure.figsize'] = 8, 6
plt.plot(valid_error)
plt.ylabel('Valid error')
plt.xlabel('Epoch #')
plt.title('model_merged_final')

set_params('./model_merged_final_best.pkl', model_merged_final)
test_loss = np.mean([test_model_merged_final(i) for i in xrange(n_test_batches)])
print('Test error: {}'.format(test_loss))
```
