```python
def save_params(file_name, model):
    with open(file_name, 'wb') as out_file:
        cPickle.dump(model.params, out_file)

def set_params(file_name, model):
    with open(file_name, 'rb') as in_file:
        new_params = cPickle.load(in_file)
    for current_p, new_p in zip(model.params, new_params):
        current_p.set_value(new_p.eval())

def reset_params(model, file_name=None):
    if file_name == None:
        file_name = 'model_initial_params.pkl'
    
    print('Before reset:')
    for p in model.params[:2]:
        print(p[0].eval()),
        
    with open(file_name, 'rb') as in_file:
        initial_params = cPickle.load(in_file)
        
    for initial, current in zip(initial_params, model.params):
        current.set_value(initial.eval())
        
    print('\nAfter reset')
    for p in model.params[:2]:
        print(p[0].eval()),
    print('\n')

def save_valid_error(file_name, valid_error):
    with open(file_name, 'wb') as out_file:
        cPickle.dump(valid_error, out_file)

def get_confidence_for_minibatch(index):
    start = index * batch_size
    end = (index + 1) * batch_size
    return np.mean(np.max(get_p_y_given_x(X[start:end]), axis=1))

def should_continue_training(min_valid_error_epoch, epoch_num, diff=25, min_epoch=100):
    if epoch_num < min_epoch:
        return True
    else:
        return False
#         return epoch_num - min_valid_error_epoch < diff
```

```python
learning_rate = 0.050
L1_reg = 0
L2_reg = 1e-4
batch_size = 100

# nkerns = (150, 75)  # default: (20, 50)
nkerns = (80, 40)
n_in = 500
temperature = 3
```

```python
# Use cifar10_david / cifar10_davidheryanto
cifar_path = os.path.join(
    os.getenv('PYLEARN2_DATA_PATH'), 'cifar10_david'
)
datasets = load_cifar10(cifar_path)

train_set_x, train_set_y = datasets[0]
valid_set_x, valid_set_y = datasets[1]
test_set_x, test_set_y = datasets[2]

# compute number of minibatches for training, validation and testing
n_train_batches = train_set_x.get_value(borrow=True).shape[0]
n_valid_batches = valid_set_x.get_value(borrow=True).shape[0]
n_test_batches = test_set_x.get_value(borrow=True).shape[0]
n_train_batches /= batch_size
n_valid_batches /= batch_size
n_test_batches /= batch_size

index = T.lscalar()  # index to a [mini]batch

x = T.matrix('x')  
y = T.ivector('y') 
```

**Sort by confidence**

```python
print '... building model'

rng = numpy.random.RandomState(23455)
random.seed(51665)

model = MLPConv(
    rng=rng,
    model_input=x,
    nkerns=nkerns, 
    n_in=n_in,
    temperature=temperature
)

cost_model = (
    model.negative_log_likelihood(y) + \
    L1_reg * model.L1 + \
    L2_reg * model.L2_sqr
)

test_model = theano.function(
    inputs=[index],
    outputs=model.errors(y),
    givens={
        x: test_set_x[index * batch_size: (index + 1) * batch_size],
        y: test_set_y[index * batch_size: (index + 1) * batch_size]
    }
)

validate_model = theano.function(
    inputs=[index],
    outputs=model.errors(y),
    givens={
        x: valid_set_x[index * batch_size: (index + 1) * batch_size],
        y: valid_set_y[index * batch_size: (index + 1) * batch_size]
    }
)

grads_model = [T.grad(cost_model, param) for param in model.params]

updates_model = [
    (param, param - learning_rate * grad)
    for param, grad in zip(model.params, grads_model)
]

train_model = theano.function(
    inputs=[index],
    outputs=cost_model,
    updates=updates_model,
    givens={
        x: train_set_x[index * batch_size: (index + 1) * batch_size],
        y: train_set_y[index * batch_size: (index + 1) * batch_size]
    }
)
```

```python
set_params('model_base_1_best.pkl', model)
valid_error = np.mean([validate_model(i) for i in xrange(n_valid_batches)])
test_error = np.mean([test_model(i) for i in xrange(n_test_batches)])
print('Valid error: {}\tTest error: {}'.format(valid_error, test_error))
```

```python
def set_avg_params(paths, model, layer_start=None, layer_end=None):
    if layer_start is None:
        layer_start = 0
    if layer_end is None:
        layer_end = len(model.params)
    
    params = []
    
    for path in paths:
        with open(path, 'rb') as in_file:
            params_loaded = cPickle.load(in_file)
            params.append(params_loaded[layer_start:layer_end])
    
    new_params = []
    for params_tuple in zip(*params):
        total = params_tuple[0].get_value()
        for param in params_tuple[1:]:
            total += param.get_value()
        avg = total / float(len(params))
        new_params.append(avg)
    
    for current_param, new_param in zip(model.params[layer_start:layer_end], new_params):
        # print(current_param.shape.eval(), new_param.shape)
        current_param.set_value(new_param)
```

```python
print '... building model_base_2'

rng = numpy.random.RandomState(23455)
random.seed(51665)

model_base_2 = MLPConv(
    rng=rng,
    model_input=x,
    nkerns=nkerns, 
    n_in=n_in,
    temperature=temperature
)
set_avg_params(['model_1_best.pkl'], model_base_2, layer_start=4)

cost_model_base_2 = (
    model_base_2.negative_log_likelihood(y) + \
    L1_reg * model_base_2.L1 + \
    L2_reg * model_base_2.L2_sqr
)

test_model_base_2 = theano.function(
    inputs=[index],
    outputs=model_base_2.errors(y),
    givens={
        x: test_set_x[index * batch_size: (index + 1) * batch_size],
        y: test_set_y[index * batch_size: (index + 1) * batch_size]
    }
)

validate_model_base_2 = theano.function(
    inputs=[index],
    outputs=model_base_2.errors(y),
    givens={
        x: valid_set_x[index * batch_size: (index + 1) * batch_size],
        y: valid_set_y[index * batch_size: (index + 1) * batch_size]
    }
)

grads_model_base_2 = [T.grad(cost_model_base_2, param) for param in model_base_2.params]

updates_model_base_2 = [
    (param, param - learning_rate * grad)
    for param, grad in zip(model_base_2.params, grads_model_base_2)
]

train_model_base_2 = theano.function(
    inputs=[index],
    outputs=cost_model_base_2,
    updates=updates_model_base_2,
    givens={
        x: train_set_x[index * batch_size: (index + 1) * batch_size],
        y: train_set_y[index * batch_size: (index + 1) * batch_size]
    }
)
```
